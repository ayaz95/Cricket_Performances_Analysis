```python
### Code to check whether table exist in the webpage


import pandas as pd
import requests
from bs4 import BeautifulSoup



def Check_table_exist(url):
    #url = f"http://www.cricmetric.com/playerstats.py?player=Jasprit+Bumrah&role=batsman&format=ODI&groupby=opp_team&playerStatsFilters=on&start_date=2021-01-01&end_date=2023-09-30&vcountry=India&start_over=0&end_over=9999&opp_team=New%20Zealand"


    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')
    else:
        print(f"Failed to retrieve the web page. Status code: {response.status_code}")
        exit()


    # Example: Find the first table element with the 'class' attribute equal to 'my-table'
    table = soup.find('table', {'class': 'my-table'})

    # Check if the table exists on the web page
    if table:
        # The table exists; you can proceed to read it using Pandas
        df = pd.read_html(str(table))
        #return True
        # df now contains the DataFrame representing the table
    else:
        print("False")#return False
        
    return 0

print(Check_table_exist(f"http://www.cricmetric.com/playerstats.py?player=Rohit+Sharma&role=batsman&format=ODI&groupby=opp_team&playerStatsFilters=on&start_date=2021-01-01&end_date=2023-09-30&vcountry=India&start_over=0&end_over=9999&opp_team=New%20Zealand"))



```

## Overall Batting Performance from 2021-23

```python
### ODI Performance Function

def odi_performance(players):
    df_odi = pd.DataFrame()
    for i in players:
        name_split = i.split()
        first_name = name_split[0]
        last_name = name_split[1]

        try:
            data_by_series = pd.read_html(f"http://www.cricmetric.com/playerstats.py?player={first_name}+{last_name}&role=all&format=ODI&groupby=player&playerStatsFilters=on&start_date=2021-01-01&end_date=2023-09-30&start_over=0&end_over=9999")
            data_by_series
            x = data_by_series[0]
            x['Player'] = i
            x['Format'] = 'ODI'
            df_odi = df_odi.append(data_by_series[0], ignore_index=True)
        except:
            continue
        
    return df_odi
```
